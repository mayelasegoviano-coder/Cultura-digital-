## Parte 3: Condicionales en Python

### Ejercicio 1
Pide la hora (0-23). Si es menor que 12 muestra 'Es mañana', si no 'Es tarde o noche'.


```python
# Escribe tu código aquí
hora=int(input("dame la hora"))
if(hora<12):
  print("es mañana")
else:
  print("es tarde")
```

### Ejercicio 2
Pide dos números y muestra cuál es mayor o si son iguales.


```python
# Escribe tu código aquí
var1=int(input("dame un numero"))
var2=int(input("dame otro numero"))
if(var1>var2):
  print("el numero es mayor")
elif(var1<var2):
  print("el numero es menor")
else:
  print("son iguales")
```

### Ejercicio 3
